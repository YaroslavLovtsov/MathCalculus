__3.__ Проверьте любым способ, является ли данные логические формулы тавтологией:

### $$a)\,\,\, (A \vee B) \rightarrow (B \vee\overline A)$$
### $$б)\,\,\, A \rightarrow (A \vee (\overline B \wedge A))$$

**Решение**

$$(A \vee B) \rightarrow (B \vee\overline A) = \overline{(A \vee B)} \vee (B \vee\overline A) = \overline A \wedge \overline B \vee B \vee \overline A = B \vee(\overline A \vee \overline A \wedge \overline B) = B \vee \overline A $$

При $A = Истина$ и $B = Ложь$ данное выражение ложно, а значит оно не является тавтологией.

$$A \rightarrow (A \vee (\overline B \wedge A)) = \overline A \vee A \vee (\overline B \wedge A) = (\overline A \vee A) \vee (\overline B \wedge A) = Истина \vee (\overline B \wedge A) = Истина $$

Данное выражение является истинным для любых $A$ и $B$, и значит является тавтологией.

__6*.__ Прочитайте высказывания, установите их истинность и постройте противоположное высказывание:

### $$a)\,\,\, \forall x\in\mathbb{R}\,\,\,\exists X\in\mathbb{R}:\,\,\, X>x;$$
### $$б)\,\,\, \forall y\in\Bigl[0; \frac{\pi}{2}\Bigr]\,\,\,\exists \varepsilon>0:\,\,\, \sin y<\sin(y+\varepsilon);$$
### $$в)\,\,\, \forall y\in\Bigl[0; \pi\Bigr)\,\,\,\exists \varepsilon>0:\,\,\, \cos y>\cos(y+\varepsilon).$$

**Решение**

а) Выражение является истинным. Фиксируем $x \in \mathbb{R} $, возьмем $X = x + 1$. Очевидно, В этом случае $X = x+1 > x$. Противоположное высказывание, соответственно ложное, записывается следующим образом
$$\exists x\in\mathbb{R}: \,\,\, \forall X\in\mathbb{R} \,\,\, X\leqslant x$$

б) Выражение не является истинным. Противоположное выражение записывается следующим образом:

$$\exists y\in\Bigl[0; \frac{\pi}{2}\Bigr]: \,\,\, \forall \varepsilon > 0 \,\,\, sin(y) \geqslant sin(y + \varepsilon) $$

Противоположное выражение является истинным. В самом деле: при $y = \frac{\pi}{2} \,\,\,  \forall \varepsilon>0 \,\,\, sin(\frac{\pi}{2}+\varepsilon) \leqslant 1 = sin(\frac{\pi}{2}) $. Таким образом, исходное выражение ложно.

в) Выражение является истинным. Фиксируем $y \in \Bigl[0; \pi\Bigr)$, возьмем $\varepsilon = \frac{\pi-y}{2}$. Очевидно, $\varepsilon > 0 $ при $y \in \Bigl[0; \pi\Bigr)$, и значит $y + \varepsilon > y$. Тогда $y + \varepsilon = \frac{\pi+y}{2} $< $\frac{\pi+\pi}{2}$ = $\pi $, то есть $y + \varepsilon  \in \Bigl[0; \pi\Bigr)$. Так как на полуоткрытом интервале $\Bigl[0; \pi\Bigr)$ функция $cos(x)$ монотонно убывает, то $cos(y + \varepsilon) = cos(\frac{\pi+y}{2}) < cos(y)$, так как $y + \varepsilon > y$. Противоположное выражение, соответственно ложное, записывается следующим образом: 

$$ \exists y \in \Bigl[0; \frac{\pi}{2}\Bigr) \,\,\, \forall \varepsilon > 0 \,\,\, cos(y) \leqslant cos(y + \varepsilon) $$
