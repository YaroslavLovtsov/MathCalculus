__1.__ Исследовать на условный экстремум функцию

### $$U=3-8x+6y,$$

если

### $$x^2+y^2=36$$

**Решение**

$$L(x,y, \lambda) = 3-8x+6y - \lambda(x^2+y^2-36)$$

$$\frac{\partial L}{\partial x} = -2\lambda x - 8, \,\,\, \frac{\partial L}{dy} = -2\lambda y + 6, \,\,\, \frac{\partial L}{\partial \lambda} = x^2+y^2 - 36$$

Приравняем частные производные к нулю и решим систему уравнений:

$$\begin{cases} \frac{\partial L}{\partial x} = 0 &&\\ \frac{\partial L}{dy} = 0 &&\\  \frac{\partial L}{\partial \lambda} = 0 \end{cases}  
\Rightarrow 
\begin{cases}  2\lambda x = - 8 &&\\ 2\lambda y = 6 &&\\  \frac{16}{\lambda^2} + \frac{9}{\lambda^2} = 36 \end{cases} $$

$$\lambda_{1,2} = \pm\frac{5}{6},\,\,\, x_{1,2} = \mp \frac{24}{5},\,\,\, y_{1,2} = \pm \frac{18}{5} $$

Найдем вторые производные и полный дифференциал второго порядка:

$$\frac{\partial^2 L}{\partial x^2} = -2\lambda,\,\,\, \frac{\partial^2 L}{\partial y^2} = -2 \lambda,\,\,\, \frac{\partial^2 L}{\partial \lambda^2} = 0$$

$$\frac{\partial^2 L}{\partial x \partial y} = 0,\,\,\, \frac{\partial^2 L}{\partial x \partial \lambda} = -2x,\,\,\, \frac{\partial^2 L}{\partial y \partial \lambda} = -2y$$

$$d^2 L = -2(\lambda dx^2 +\lambda dy^2 +2xdxd\lambda + 2ydyd\lambda)$$

Продифференцируем обе части нашего условия $ x^2+y^2 = 36$:

$$2xdx + 2ydy = 0 \Rightarrow 2xdxd\lambda + 2ydyd\lambda = 0$$.

Значит, $d^2 L = -2\lambda(dx^2 + dy^2)$. Рассмотрим $d^2 L$ в наших стационарных точках:

1) $\lambda = \frac{5}{6},\,\,\, x = - \frac{24}{5},\,\,\, y = \frac{18}{5}$:

$$d^2 L = -\frac{5}{3} (dx^2 + dy^2) \leqslant 0 \,\,\, \forall dx, dy $$

Точка $(-\frac{24}{5},\frac{18}{5})$ является точкой максимума.

2) $\lambda = -\frac{5}{6},\,\,\, x = \frac{24}{5},\,\,\, y = - \frac{18}{5}$:

$$d^2 L = \frac{5}{3} (dx^2 + dy^2) \geqslant 0 \,\,\, \forall dx, dy $$

Точка $(\frac{24}{5},-\frac{18}{5})$ является точкой минимума.

__2.__ Исследовать на условный экстремум функцию

### $$U=2x^2+12xy+32y^2+15,$$

Приравняем частные производные к нулю и решим систему уравнений:

$$\begin{cases} \frac{\partial L}{\partial x} = 0 &&\\ \frac{\partial L}{dy} = 0 &&\\  \frac{\partial L}{\partial \lambda} = 0 \end{cases}  
\Rightarrow 
\begin{cases}  2(2-\lambda)x + 12y = 0 &&\\ 12x+32(2-\lambda)y = 0 &&\\  x^2+16y^2 = 64 \end{cases} $$

Найдем при каких $\lambda$ система из первых двух уравнений имеет нетривиальное решение. Найдем вторые производные и второй дифференциал:

$$\frac{\partial^2 L}{\partial x^2} = 4 - 2\lambda,\,\,\,\frac{\partial^2 L}{\partial y^2} = 64 - 32\lambda,\,\,\,\frac{\partial^2 L}{\partial \lambda^2} = 0$$


$$\frac{\partial^2 L}{\partial x \partial y} = 0,\,\,\,\frac{\partial^2 L}{\partial x \partial \lambda} = -2x,\,\,\,\frac{\partial^2 L}{\partial y \partial \lambda} = -32y$$

$$d^2 L = (2-\lambda)(2dx^2+32dy^2)-2x\,dx\,d\lambda - 32y\,dy\,d\lambda$$

Продифференцируем обе части нашего условия $x^2 + 16y^2 = 64: 2xdx+32ydy = 0$

Тогда $-2x\,dx\,d\lambda - 32y\,dy\,d\lambda = 0,\,\,\, d^2 L = (2-\lambda)(2dx^2+32dy^2)$

$d^2L = -\frac{3}{2}(dx^2+dy^2) \leqslant 0 \,\,\, \forall dx, dy$ в обеих найденных точках.

__6**.__ Найти все корни уравнения:
    
### $$f(x)=g(x)$$

на участке $[0, 20]$

Где $y_{1}=f(x)$ и $y_{2}=g(x)$ - функции из предыдущего решения.

**Решение**

Построение графика и поиск численного решения приведены ниже. Аналитическое решение дифференциального уравнения показало, что:

$$f(x) = 1 - \cos x +\frac{1}{2}\sin x$$
$$g(x) = \cos x + 2 \sin x$$

Решением уравнения $f(x) = g(x)$ являются:

$$x = \pm \arccos\frac{2}{5} + \arccos\frac{4}{5} + 2\pi k, k \in \mathbb Z$$

```
from matplotlib import pylab as plt
import numpy as np
%matplotlib inline

step = 355 / 678 / 450000

x_curr = 0
f_curr = 0
g_curr = 1
f_primes_curr = 1 / 2
g_primes_curr = 2

x = [x_curr]
f = [f_curr]
g = [g_curr]
f_primes = [f_primes_curr]
g_primes = [g_primes_curr]

x_print = [x_curr]
f_print = [f_curr]
g_print = [g_curr]

count = 0

while x_curr <= 20:

  count += 1
  x_curr += step
  g_curr += step * g_primes_curr
  f_curr += step * f_primes_curr
  f_primes_curr = g_curr / 2
  g_primes_curr = 2 - 2 * f_curr

  x.append(x_curr)
  f.append(f_curr)
  g.append(g_curr)
  f_primes.append(f_primes_curr)
  g_primes.append(g_primes_curr)

  if count % 100 == 0:
    x_print.append(x_curr)
    f_print.append(f_curr)
    g_print.append(g_curr)
  
plt.figure(figsize=(6,6))
plt.axis([0, 20, -3, 3])

x_print = np.array(x_print)

plt.plot(x_print,np.array(f_print), c='b')
plt.plot(x_print,np.array(g_print), c='r')
plt.show()
```
